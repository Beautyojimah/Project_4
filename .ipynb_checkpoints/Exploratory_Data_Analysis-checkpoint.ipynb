```python
#import dependencies
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
#Read in data into Dataframe
merged_df = pd.read_csv("Resources/cleaned_merged_df.csv")

merged_df.head()
```

### What is the sales distribution?

```python
# Visualise the distribution of Sales
plt.figure(figsize=(8, 6))
sns.histplot(merged_df[merged_df['Sales'] > 0]['Sales'], kde=True,bins=30)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

### Correlation Analysis

```python
# plot a correlation matrix of the variables to understand their relationship
# observation: There is a strong positive correlation (0.89) between Sales and Customers. 
# This suggests that stores with more customers tend to have higher sales, which is a logical assumption.

# The Promo variable has a weak positive correlation with both Sales (0.45) and Customers (0.32). 
# This indicates that on days when a store runs a promotion, it tends to have higher sales and attract more customers.

# The Open variable, which indicates whether a store is open (1) or closed (0), has a strong positive correlation 
# with Sales (0.68) and Customers (0.62). This is expected as stores would have sales and customers only when they are open.

# Promo2 does not show a significant correlation with Sales or Customers. 
# This suggests that the continuous promotion (Promo2) does not have a strong linear impact on sales or the number of customers.

# DayOfWeek has a small negative correlation with Sales and Customers. This might 
# suggest that sales and customer traffic are slightly lower later in the week (explore this further).


# Calculate the correlation matrix
correlation_matrix = merged_df.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

### What is the monthly sales across different years?

```python
# Observations: there appears to be a clear seasonal trend/pattern in sales. 
# Sales tend to peak around June/July and December.

# Sales vary across different years, but the overall seasonal pattern seems consistent. 
# This suggests that while the total sales might change, the shopping behavior in terms of seasonality remains similar.

# There are noticeable fluctuations in sales from month to month.

# Convert the 'Date' column to datetime format
merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Extract year and month from the 'Date' column
merged_df['Year'] = merged_df['Date'].dt.year
merged_df['Month'] = merged_df['Date'].dt.month_name()

# Group by 'Year' and 'Month' and calculate the total sales for each month of each year
monthly_sales = merged_df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Sort the data by 'Year' and 'Month' for proper visualisation
monthly_sales['Month'] = pd.Categorical(monthly_sales['Month'], categories=
    ['January','February','March','April','May','June','July','August','September','October','November','December'],
    ordered=True)
monthly_sales = monthly_sales.sort_values(by=['Year', 'Month'])

# Plotting the monthly sales across different years
plt.figure(figsize=(16, 8))
sns.lineplot(data=monthly_sales, x='Month', y='Sales', hue='Year', marker='o', palette='tab10')
plt.title('Monthly Sales Across Different Years')
plt.ylabel('Total Sales')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()

plt.show()
```

### What is the average sale per week

```python
# observation: it gives insight into why the dayofweek has small negative correlation 
#with sales as the sales drop by each day and the weekend has significantly less sales.

#Extract the day of the week and month from the 'Date' column
merged_df['DayOfWeek'] = merged_df['Date'].dt.day_name()

# Group the merged dataset by 'DayOfWeek' and calculate the average sales for each day of the week
average_sales_per_day = merged_df.groupby('DayOfWeek')['Sales'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]).reset_index()

# Initialise the figure with two subplots
plt.figure(figsize=(8, 6))

# Plot the average sales for each day of the week
sns.barplot(x='DayOfWeek', y='Sales', data=average_sales_per_day, palette='viridis')
plt.title('Average Sales for Each Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Sales')

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```

### What is the total sales for different store types.

```python
# Analyse the sales per StoreType
store_type_sales = merged_df.groupby('StoreType')['Sales'].mean().reset_index()
sns.barplot(x='StoreType', y='Sales', data=store_type_sales)
plt.title('Average Sales by Store Type')

plt.show()
```

### What is the avearge sales by assortment levels?

- In retail, "assortment level" refers to the variety and range of products that a store offers to its customers. It encompasses the diversity of products, brands, categories, and variations available for purchase. This dataset contains three assortment levels (a = basic, b = extra, c = extended)

- A basic assortment level refers to a retailer offering a limited range of products, focusing primarily on essential or core items. This approach is commonly used by smaller stores, discount stores, or convenience stores where the emphasis is on meeting the immediate and basic needs of customers.

- An extra assortment level means that, in addition to the basic assortment, the retailer offers additional variations, brands, or product categories. This could include seasonal items, special editions, or complementary products.

- An extended assortment level represents the broadest and most diverse range of products offered by a retailer. This includes a wide variety of product categories, brands, variations, and niche products. Morelike a one-stop shop.

```python
# group the data by assortment level and calculate the average sales for each level (a = basic, b = extra, c = extended).

#observation: We observe variations in average sales across different assortment levels.

# Group the merged_df dataframe by 'Assortment' and calculate the average sales for each assortment level
average_sales_by_assortment = merged_df.groupby('Assortment')['Sales'].mean().reset_index()

# Initialise the figure
plt.figure(figsize=(10, 6))

# Plot the average sales for each assortment level
sns.barplot(x='Assortment', y='Sales', data=average_sales_by_assortment, palette='viridis')

# Set the title and labels of the plot
plt.title('Average Sales for Each Assortment Level')
plt.xlabel('Assortment Level')
plt.ylabel('Average Sales')

# Display the plot
plt.show()
```

###  How do sales differ by store type and assortment level?

```python
## group the data by store type and assortment level to analyse variations in sales.

#obsevration: From the plots, we can observe variations in both average sales 
#and the average number of customers across different store types and assortment levels.


average_sales_by_type_assortment=merged_df.copy()

# Group the data by 'StoreType' and 'Assortment' and calculate the average sales and average number of customers
average_sales_by_type_assortment = average_sales_by_type_assortment.groupby(['StoreType', 'Assortment'])['Sales'].mean().reset_index()

# Initialise the figure with two subplots
plt.figure(figsize=(8, 6))

# Plot the average sales by store type and assortment level
sns.barplot(x='StoreType', y='Sales', hue='Assortment', data=average_sales_by_type_assortment, palette='viridis')
plt.title('Average Sales by Store Type and Assortment Level')
plt.xlabel('Store Type')
plt.ylabel('Average Sales')

# Adjust the layout and display the plots
plt.tight_layout()
plt.show()
```

### Average sales on School Holiday vs Non-school Holiday

```python
#Impact of school holidays on average sales by comparing the average sales on 
#school holiday days (SchoolHoliday == 1) with the average sales on non-school holiday days (SchoolHoliday == 0)

#observations: From the plot, we observe that the average sales on school holiday days are slightly
# higher than on non-school holiday days, indicating that school holidays have a positive impact on sales.



## Perform aggregation to calculate the average sales on school holiday days and non-school holiday days
average_sales_school_holiday = merged_df.groupby('SchoolHoliday')['Sales'].mean().reset_index()
average_sales_school_holiday['SchoolHoliday'] = average_sales_school_holiday['SchoolHoliday'].map({0: 'Non-School Holiday', 1: 'School Holiday'})

# Initialise the figure
plt.figure(figsize=(8, 6))

# Create a bar plot of average sales by school holiday status
sns.barplot(x='SchoolHoliday', y='Sales', data=average_sales_school_holiday, palette='viridis')

# Set the title and labels of the plot
plt.title('Average Sales on School Holiday vs Non-School Holiday Days')
plt.xlabel('Day Type')
plt.ylabel('Average Sales')

# Display the plot
plt.show()
```

### Average sales on State Holiday vs Non-state Holiday

```python
#To analyse how sales vary during state holidays and school holidays compared to regular days.
#Compare  the average sales on state holiday days (StateHoliday != '0') with the average sales on 
#non-holiday days (StateHoliday == '0'). After that, analyse the impact of school holidays on average sales.

#Observations: From the plot, we observe that the average sales on state holiday days 
# are lower than on non-holiday days, indicating that state holidays impact sales negatively.


# Calculate the average sales on state holiday days and non-holiday days
average_sales_non_holiday = merged_df.loc[merged_df['StateHoliday'] == '0', 'Sales'].mean()
average_sales_state_holiday = merged_df.loc[merged_df['StateHoliday'] != '0', 'Sales'].mean()

# Create a DataFrame to hold the results
average_sales_state_holiday_df = pd.DataFrame({
    'Day Type': ['Non-Holiday', 'State Holiday'],
    'Average Sales': [average_sales_non_holiday, average_sales_state_holiday]
})

# Initialise the figure
plt.figure(figsize=(8, 6))

# Create a bar plot of average sales by state holiday status
sns.barplot(x='Day Type', y='Average Sales', data=average_sales_state_holiday_df, palette='viridis')

# Set the title and labels of the plot
plt.title('Average Sales on State Holiday vs Non-Holiday Days')
plt.xlabel('Day Type')
plt.ylabel('Average Sales')

# Display the plot
plt.show()
```

### Comparison of Sales on School Holiday and State Holiday

```python
#observation: average sales on school holiday days are slightly, higher that stateholiday sales.

# Filtering out the rows where the store is closed
open_data = merged_df[merged_df['Open'] == 1]

# Calculating total sales during school holidays, state holidays, and regular days
sales_school_holiday = open_data[open_data['SchoolHoliday'] == 1]['Sales'].sum()
sales_state_holiday = open_data[open_data['StateHoliday'] != '0']['Sales'].sum()

# Creating a DataFrame for visualisation
sales_comparison = pd.DataFrame({
    'Holiday Type': ['School Holiday', 'State Holiday'],
    'Total Sales': [sales_school_holiday, sales_state_holiday]
})

# Visualising the total sales comparison
plt.figure(figsize=(8, 6))
sns.barplot(data=sales_comparison, x='Holiday Type', y='Total Sales', palette='viridis')
plt.title('Total Sales Comparison: School Holiday vs State Holiday')
plt.ylabel('Total Sales')
plt.xlabel('Holiday Type')
plt.tight_layout()
plt.show()
```

### Does promotion have an impact on sales?

```python
#observation: promotion impacts sales

# Group the data by 'Promo' and calculate the average sales and customers for each group
promo_sales_customers = merged_df.groupby('Promo')[['Sales', 'Customers']].mean().reset_index()

# Define the promo labels
promo_labels = ['No Promo', 'Promo']

# Plot the average sales and customers for each promo group
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for average sales
ax1.bar(promo_sales_customers['Promo'], promo_sales_customers['Sales'], 
        alpha=0.7, label='Average Sales', color='dodgerblue')
ax1.set_xlabel('Promotion')
ax1.set_ylabel('Average Sales')
ax1.set_title('Impact of Promotion on Average Sales and Customer Traffic')
ax1.set_xticks(promo_sales_customers['Promo'])
ax1.set_xticklabels(promo_labels)

# Line plot for average customers (using twinx to create a second y-axis)
ax2 = ax1.twinx()
ax2.plot(promo_sales_customers['Promo'], promo_sales_customers['Customers'], 
         marker='o', linestyle='-', color='darkorange', label='Average Customers')
ax2.set_ylabel('Average Customers')

# Show legend for both y-axes
fig.legend(loc="upper left", bbox_to_anchor=(0,1), bbox_transform=ax1.transAxes)

plt.show()
```

### Do stores participating in Promo2 have higher sales than those that do not?

```python
#compare the average sales of stores participating in Promo2 with those that do not participate.

#observation: From the plot, we observe that the average sales of stores participating in Promo2 
# are slightly lower than those of stores not participating in Promo2.

#so Participating in Promo2 does not seem to have a significant positive impact on the average sales of the stores.
#This insight can be valuable in evaluating the effectiveness of Promo2 and making decsions on promotional strategies

# Group the merged dataset by 'Store' and calculate the average sales for each store
average_sales_per_store = merged_df.groupby('Store')['Sales'].mean().reset_index()


# Calculate the average sales for stores participating in Promo2 and those not participating
average_sales_by_promo2 = merged_df.groupby('Promo2')['Sales'].mean().reset_index()
average_sales_by_promo2['Promo2'] = average_sales_by_promo2['Promo2'].map({0: 'No Promo2', 1: 'Promo2'})

# Initialise the figure
plt.figure(figsize=(8, 6))

# Create a bar plot of average sales by Promo2 participation status
sns.barplot(x='Promo2', y='Sales', data=average_sales_by_promo2, palette='viridis')

# Set the title and labels of the plot
plt.title('Average Sales by Promo2 Participation Status')
plt.xlabel('Promo2 Participation Status')
plt.ylabel('Average Sales')

# Display the plot
plt.show()
```

### Compare the sales for promo and promo2

```python
#To compare the sales for promo and promo2
#observations: When "Promo" is not running (Promo = 0), the stores not participating in "Promo2"
#have higher average sales than the stores that are participating in "Promo2".

#Similarly, when "Promo" is running (Promo = 1), the stores not participating in "Promo2" 
#also have higher average sales,though the difference is less pronounced, as compared to the stores 
#that are participating in "Promo2".

#it therefore, suggests that on average, 
#stores not participating in "Promo2" tend to have higher sales regardless of whether "Promo" is running or not.
#However, the difference in average sales is smaller when "Promo" is active. This could imply that while "Promo" 
#has a positive effect on sales, 
#"Promo2" does not consistently contribute to an increase in sales. 


# A key question is why do sales reduce for stores that integrate promo 1 and 2 if promo1 gives the highest sales.
# it may be due to resource allocation and requires further exploration

# Group the data by 'Promo' and 'Promo2' and calculate the average sales for each group
promo_comparison = merged_df.groupby(['Promo', 'Promo2'])['Sales'].mean().reset_index()

# Pivot the DataFrame for better visualisation
promo_comparison_pivot = promo_comparison.pivot(index='Promo', columns='Promo2', values='Sales').rename(columns={0: 'No Promo2', 1: 'Promo2'})

# Plot the comparison of average sales for Promo and Promo2
promo_comparison_pivot.plot(kind='bar', figsize=(10, 6), color=['#1f77b4', '#ff7f0e'], alpha=0.7)
plt.title('Comparison of Average Sales for Promo and Promo2')
plt.xlabel('Promo')
plt.ylabel('Average Sales')
plt.xticks(ticks=[0, 1], labels=['No Promo', 'Promo'], rotation=0)
plt.legend(title='Promo2')
plt.show()
```
